## import

```python
import requests
import re
from collections import Counter
```

```python
with open("resouce/alice29.txt", "r", encoding="utf-8") as f:
    data = f.read()
data[:500]
```

## Clean Text

```python
cleaned = data.lower()
cleaned = re.sub(r"[^a-z\s]", " ", cleaned)
cleaned = re.sub(r"\s+", " ", cleaned).strip()
cleaned[:500]
```

## Tokenize Sentence and Word

```python
sentences = re.split(r'(?<=[.!?])\s+', data)
sentences = [s.strip() for s in sentences if len(s.strip()) > 0]

sentence_cleaned = [s.lower() for s in sentences]
print(len(sentences))
sentences[:10]
```
